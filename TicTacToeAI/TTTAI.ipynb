# Tic Tac Toe AI Using Neural Networks

This document describes the design and implementation of an artificial intelligence with the purpose of playing the game "Tic Tac Toe" against either another instance of the same AI or other players, for example humans.

The project is part of our seventh-semester course "Data Mining and Machine Learning".

In the following texts and code examples we want to explain our design decisions and how our implementation works.

## Basic design

At the beginning of the project, we researched several types of learning algorithms in order to find one that might be suitable for our purposes. After a hint from our lecturers to look into Qlearning, we quickly decided that we wanted to apply an algorithm based around rewards and punishments. These types of algorithms are called "reinforcement learning algorithms" and work basically by carrying out only random actions when first starting to learn and observing the results. Actions that yield positive results are rewarded with points, actions that lead to failure are punished by taking away points. Blocking an enemy win for example could result in 400 points while the punishment for losing a game is -500 points.

The algorithm keeps track of all actions it carried out and the total sum of points they resulted in. At later stages, it can then refer to these records to decide which action is most likely to lead to the desired result, i.e. winning the game.

The graphic below illustrates the action/ feedback loop between the game AI ("Agent") and the game manager that controls the game status ("Environment"):  
  
  


Very simply put, the Environment supplies the agent with information about the current state (in our case, the tic tac toe field and which segments have been set by which party). Based on this information, the Agent decides on an action to take and tells the Environment about it. After evaluating the action, the Environment passes the new state as well as the earned reward to the Agent. This loop continues until a certain condition is met (until the game is either won by one of the parties or results in a draw).
